
# What does the customer rating look like and is it skewed?(Use normal distribution plot)

# Is there any difference in aggregate sales across branches?(Use bar graph)

```python
branch_sales=data.groupby('Branch')['Total'].sum()

plt.bar(branch_sales.index,branch_sales.values,width=0.5,color='blue')
plt.title('Total Sales by Branches')
plt.ylabel("Sales_Total")
plt.xlabel("Branches")
```

# the most popular payment method used by customers?(Use bar graph)

```python
payment_method_count=data['Payment'].value_counts()
payment_method_count=payment_method_count.sort_values(ascending=False)
plt.bar(payment_method_count.index,payment_method_count.values,width=0.5,color='blue',label=True)
plt.title('Payment_method_count')
plt.xlabel("payment method")
plt.ylabel("count of customers")
```
